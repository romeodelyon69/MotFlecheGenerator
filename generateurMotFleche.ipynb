```python
class GrilleMotsFlechés:
    def __init__(self, lignes, colonnes):
        self.lignes = lignes
        self.colonnes = colonnes
        self.grid = [[[0,0,None, None] for _ in range(colonnes)] for _ in range(lignes)]                                 #un élément [5, 4, Nod(a,...), Nod(a,...)] veut dire que la case contient la lettre a dans un mot de 5 lettres horizontalement et 4 lettres verticalement 

    def draw(self):
        for i in range(self.lignes):
            for j in range(self.colonnes):
                if(self.grid[i][j][2] != None and self.grid[i][j][0] != 0):
                    print(self.grid[i][j][2].e + '|', end='')
                else:
                    print('¤|', end = '')
            print()
        
        print()
        print()
  
    
    def add_def(self, definition, ligne, colonne):
        if self.grid[ligne][colonne] == '':
            self.grid[ligne][colonne] = definition
        
        else:
            self.grid[ligne][colonne].texte.append(definition.texte[0])
            self.grid[ligne][colonne].direction.append(definition.direction[0])

    def add_letter(self, pos, nod_h, nod_v):
        self.grid[pos[0]][pos[1]][2] = nod_h
        self.grid[pos[0]][pos[1]][3] = nod_v
    
    def remove_letter(self, pos):
        self.grid[pos[0]][pos[1]][2] = None
        self.grid[pos[0]][pos[1]][3] = None
    
    def next_pos(self, pos):
        l = pos[0]
        c = pos[1]

        if c+1 < self.colonnes:
            return (l, c+1)
        
        return (l+1, 0)
```

```python
class Definition:
    def __init__(self, texte, direction):
        self.texte = texte                                          #ce sont des listes car il peut y a voir deux définitions dans une même case 
        self.direction = direction                                  #liste aussi
                                                                    
```

```python
#e représent un character (ou bien le caractère vide également)
class Node:
    def __init__(self, e, sons):
        self.e = e 
        self.sons = sons

    def add_node(self, e):
        if not e in self.sons:
            self.sons[e] = Node(e, {})
        return self.sons[e]
    
    def add_word(self, word):
        if word == '':
            return True
        subTree = self.add_node(word[0])
        subTree.add_word(word[1:])
    
    def print(self):
        print(self.e)

        for fils in self.sons.values():
            fils.print()
            

        

def dico2tree(word_list):
    tree = Node('', {})
    for word in word_list :
        tree.add_word(word)
    return tree



```

```python
def init_trees():
    with open('fr_small.txt', 'r') as fichier:
        lignes = fichier.readlines()

    word_list = [ligne.strip() for ligne in lignes]

    length_list = [len(word) for word in word_list]

    max_length = max(length_list)

    word_lists = [[] for _ in range(max_length+1)]

    for i in range(len(word_list)):
        word_lists[length_list[i]].append(word_list[i])

    trees = [dico2tree(dico) for dico in word_lists]

    return trees
```

```python
dim_grid = 6,6
def init_grid(dim_grid, trees, black_cell_grid):
    mot_fleche = GrilleMotsFlechés(dim_grid[0], dim_grid[1])
    grid = mot_fleche.grid

    #on rentre dans chaque case la taille du mot restant à partir de cette case, une pemière fois horizontalement puis verticalment 
    counter = 0
    for i in range(dim_grid[0]-1, -1, -1):
        counter = 0
        for j in range(dim_grid[1]-1, -1, -1):
            counter += 1
            if(black_cell_grid[i][j]):
                counter = 0
            grid[i][j][0] = counter
    
    counter = 0
    for j in range(dim_grid[1]-1, -1, -1):
        counter = 0
        for i in range(dim_grid[0]-1, -1, -1):
            counter += 1
            if(black_cell_grid[i][j]):
                counter = 0
            grid[i][j][1] = counter

    #on va maintenant remplir le contenu des cases noirs (cases de déf)
    for i in range(dim_grid[0]):
        for j in range(dim_grid[1]):
            if(black_cell_grid[i][j]):
                
                if(j < dim_grid[1] - 1):
                    grid[i][j][2] = trees[grid[i][j+1][0]]
                if(i < dim_grid[0] - 1):
                    grid[i][j][3] = trees[grid[i+1][j][1]]

    return mot_fleche
```

```python

def complete_grid(mot_fleche, pos, trees):
    if pos == (mot_fleche.lignes, 0):
        mot_fleche.draw()
        return
    
    next_pos = mot_fleche.next_pos(pos)

    verifX = mot_fleche.grid[pos[0]][pos[1]][0]
    verifY = mot_fleche.grid[pos[0]][pos[1]][1]
    
    if mot_fleche.grid[pos[0]][pos[1]][0] == 0 and mot_fleche.grid[pos[0]][pos[1]][1] == 0:                                   #il s'agit d'une case contenant une défintion, on ne la touche pas
        complete_grid(mot_fleche, next_pos, trees)
    
    else:
        
        if pos[0] == 0:                                                                                    #on récupère ici les deux arbres relatifs à la position de la case dans les mots courants, bien faire attention à mettre les bons arbres dans les cases des defs à l'initialisation
            tree_v = trees[mot_fleche.grid[pos[0]][pos[1]][1]]
        else:
            tree_v = mot_fleche.grid[pos[0]-1][pos[1]][3]

        if pos[1] == 0:
            tree_h = trees[mot_fleche.grid[pos[0]][pos[1]][0]]
        else:
            tree_h = mot_fleche.grid[pos[0]][pos[1]-1][2]

        # On va maintenant récupérer l'intesection des deux listes de lettres possibles afin de tester toutes les possibilités qui nous sont offertes 

        poss_v = set(tree_v.sons.keys())
        poss_h = set(tree_h.sons.keys())

        poss = poss_v & poss_h
        poss_list = list(poss)
        random.shuffle(poss_list)

        for scenario in poss_list:
            mot_fleche.add_letter(pos, tree_h.sons[scenario], tree_v.sons[scenario])
            varTest1 = tree_h.sons[scenario].sons
            varTest2 = tree_v.sons[scenario].sons
            complete_grid(mot_fleche, next_pos, trees)
            mot_fleche.remove_letter(pos)





```
